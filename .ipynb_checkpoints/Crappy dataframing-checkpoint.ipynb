```python
import pandas as pd
import json
import os
```

```python
l = []

for root, dirs, files in os.walk(".", topdown=False):
    for name in files:
        if name.endswith('.json'):
            l.append(str(os.path.join(name)))
```

```python
l
```

```python
df = pd.read_json(l[1], lines=True)
# drop columns
columns_to_drop = ['id_str', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'quoted_status_id_str',
                           'favorited', 'retweeted', 'filter_level', 'timestamp_ms', 'retweeted_status', 'delete']
df.drop(columns_to_drop, axis=1, inplace=True)
# print the modified dataframe
df_user= df[['user']].copy()
df_user
```

```python
s = []
s2 = []
dict_user = df_user.to_dict()
for d in dict_user.values():
    for n in d.values():
        s.append(n)

for g in s[0]:
    s2.append(g)
df_userok = pd.DataFrame.from_dict(s[0], orient='index')
```
