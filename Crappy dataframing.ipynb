```python
import pandas as pd
import json
import os
import sqlite3
```

```python
l = []

for root, dirs, files in os.walk(".", topdown=False):
    for name in files:
        if name.endswith('.json'):
            l.append(str(os.path.join(name)))
```

```python
df = pd.read_json(l[1])
# drop columns
columns_to_drop = ['id_str', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'quoted_status_id_str',
                           'favorited', 'retweeted', 'filter_level', 'timestamp_ms', 'retweeted_status', 'delete']
df.drop(columns_to_drop, axis=1, inplace=True)
# print the modified dataframe

list_of_user_values = df['user'].tolist()
list_of_user_names = [n for n in list_of_user_values[0]]
df_user_separate_1 = pd.DataFrame([list_of_user_values[0]])
df_user_separate_1
```

```python
df = pd.read_json(l[0], lines=True)
# drop columns
columns_to_drop = ['id_str', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'quoted_status_id_str',
                           'favorited', 'retweeted', 'filter_level', 'timestamp_ms', 'retweeted_status', 'delete']
df.drop(columns_to_drop, axis=1, inplace=True)
# print the modified dataframe

list_of_user_values = df['user'].tolist()
list_of_user_names = [n for n in list_of_user_values[0]]
df_user_separate_2 = pd.DataFrame([list_of_user_values[0]])
df_user_separate_2
```
