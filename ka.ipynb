```python
import os
import json
import sqlite3
import pandas as pd

# Establish a SQLite connection
conn = sqlite3.connect('tweets_database.db')

def load_json_to_sqlite():
    # Initialize counters
    success_count = 0
    fail_count = 0
    json_fail_count = 0

    # Specify the directory where JSON files are located
    directory = 'data/'

    # List of columns to keep
    cols_to_keep = ['created_at', 'id', 'text', 'truncated', 'in_reply_to_status_id',
       'in_reply_to_user_id', 'is_quote_status', 'quote_count', 'reply_count',
       'retweet_count', 'favorite_count', 'lang', 'quoted_status_id',
       'possibly_sensitive', 'user_created_at', 'user_id', 'user_screen_name',
       'user_location', 'user_description', 'user_protected', 'user_verified',
       'user_followers_count', 'user_friends_count', 'user_favourites_count',
       'user_statuses_count', 'full_text', 'qt_created_at', 'qt_id', 'qt_text',
       'qt_in_reply_to_user_id', 'qt_in_reply_to_screen_name',
       'qt_contributors', 'qt_is_quote_status', 'qt_quote_count',
       'qt_reply_count', 'qt_retweet_count', 'qt_favourite_count',
       'qt_favorited', 'qt_retweeted', 'qt_possibly_sensitive', 'qt_lang',
       'qt_full_text', 'qt_user_created_at', 'qt_user_id',
       'qt_user_screen_name', 'qt_user_location', 'qt_user_description',
       'qt_user_protected', 'qt_user_verified', 'qt_user_followers_count',
       'qt_user_friends_count', 'qt_user_favourites_count',
       'qt_user_statuses_count']

    # Load each JSON file
    for filename in os.listdir(directory):
        if filename.endswith('.json'):
            try:
                # Use pandas read_json with lines=True
                df = pd.read_json(os.path.join(directory, filename), lines=True)
            except ValueError:
                json_fail_count += 1
                continue
            
            # Check which columns to keep are in the DataFrame
            cols_to_use = [col for col in cols_to_keep if col in df.columns]
            
            # If no matching columns, skip this file
            if not cols_to_use:
                fail_count += 1
                continue
            
            # Keep only the columns specified
            df = df[cols_to_use]
            
            # Write the dataframe to SQLite database
            df.to_sql('tweets', conn, if_exists='append', index=False)
            
            # Increment success counter
            success_count += 1

    print(f'Successfully processed {success_count} files.')
    print(f'Failed to process {fail_count} files due to missing columns.')
    print(f'Failed to process {json_fail_count} files due to JSON decoding errors.')

# Call the function
load_json_to_sqlite()

# Close the SQLite connection
conn.close()
```
